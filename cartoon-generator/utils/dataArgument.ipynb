```python
import cv2
import matplotlib.pyplot as plt # plt 用于显示图片
import imutils
import random

def rotate(img,name,i,flip, center, angle, new_dim):
    if(flip==1):
        img = cv2.flip(img, 1)
    M = cv2.getRotationMatrix2D(center=center,angle=angle, scale=1.0)
    img = cv2.warpAffine(img, M, new_dim)
    cv2.imwrite(name+'_rotate_'+ str(i)+'.png', img)


def AffineTrans(img,name,i,flip,pts2):
    if(flip==1):
        img = cv2.flip(img, 1)
    rows, cols, _ = img.shape
    pts1 = np.float32([[20, 20], [30, 20], [20, 0]])  # 源图像中的三角形顶点坐标
    # pts2 = np.float32([[20, 20], [30, 20], [25, 0]])  # 目标图像中的三角形顶点坐标
    M = cv2.getAffineTransform(pts1, pts2)  # 计算出仿射变换矩阵
    dst = cv2.warpAffine(img, M, (cols, rows))  # 应用仿射变换
    cv2.imwrite(name+'_affine_'+ str(i)+'.png', dst)

def PerspectiveTrans(img,name,i,flip,pts2):
    if(flip==1):
        img = cv2.flip(img, 1)
    rows, cols, _ = img.shape
    pts1 = np.float32([[20, 20], [30, 20], [20, 0],[30,0]])  # 源图像中的三角形顶点坐标
    pts2 = np.float32([[20, 20], [30, 20], [21, 0],[30,0]])  # 目标图像中的三角形顶点坐标
    M = cv2.getPerspectiveTransform(pts1, pts2)  # 计算出仿射变换矩阵
    dst = cv2.warpPerspective(img, M, (cols, rows))  # 应用仿射变换
    cv2.imwrite(name+'_perspective_'+ str(i)+'.png', dst)


name = '10044740'
image = cv2.imread(name+"/img.png")
label = cv2.imread(name+"/label.png")
height, width, channel = image.shape
cx = int(width / 2)
cy = int(height / 2)
center = (cx, cy)
new_dim = (width, height)


#旋转
# for i in range(5):
#     flip = random.randint(0,1)==1
#     angle = random.randint(-59,59)
#     rotate(image, "image_"+name ,i,flip, center, angle, new_dim)
#     rotate(label, "label_"+name ,i,flip, center, angle, new_dim)

#仿射变换
# for i in range(5):  
#     flip = random.randint(0,1)==1
#     pts2 = np.float32([[20,20],[30,random.randint(16,24)],[random.randint(16,24),0]])
#     AffineTrans(image, "image_"+name,i,flip,pts2)
#     AffineTrans(label, "label_"+name,i,flip,pts2)

PerspectiveTrans(image,"image_"+name,3,0,[])
    

#透视变换




# b,g,r = cv2.split(img) 
# img2 = cv2.merge([r,g,b]) 
# plt.imshow(img2) 
# plt.show()
```
